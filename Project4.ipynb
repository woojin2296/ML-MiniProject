## 데이터셋 입력 및 라벨 분리

#### 데이터셋 불러오기


```python
import numpy as np
import pandas as pd

input_file_name = "data/winequality-red.csv"
df_train = pd.read_csv(input_file_name)
```

#### 데이터셋 확인

```python
df_train.head()
```

```python
df_train['quality'].value_counts()
```

```python
df_train.describe()
```

```python
df_train.isnull().sum()
```

```python
for column in df_train.columns:
    mean = df_train[column].mean()
    std = df_train[column].std()
    outliers = df_train[(df_train[column] < mean - 3*std) | (df_train[column] > mean + 3*std)]
    print(f'{column} 이상치 개수: {len(outliers)}')
```

#### DataFrame 객체를 numpy 배열로 변환


```python
np_data = df_train.to_numpy()
np_data
```

#### 학습 셋과 라벨 분리

```python
x_train = np_data[:, 1:]
# x_train
```

```python
y_train = np_data[:, 0]
y_train
```

## 데이터 전처리

#### Normalization

```python
x_train = x_train / 255.0
```

#### label One Hot Encode

```python
from sklearn.preprocessing import OneHotEncoder

OH_encoder = OneHotEncoder()
y_train_res = y_train.reshape(-1, 1)
y_train_encoded = OH_encoder.fit_transform(y_train_res).toarray()
```

## 확률적 경사 하강법 분류기

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(max_iter=5, tol=None, random_state=42)
sgd_clf.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, x_train, y_train, cv=10, scoring="accuracy")
```
